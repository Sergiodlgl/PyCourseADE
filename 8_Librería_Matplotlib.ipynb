# MÓDULO 8: LIBRERÍA MATPLOTLIB

Matplotlib es una librería muy utilizada en Python para la creación de gráficos y visualizaciones. Permite representar datos de forma visual, lo cual es especialmente útil para análisis económico-financiero, presentaciones o informes. En este módulo veremos cómo generar gráficos básicos como líneas, barras e histogramas, y cómo añadir etiquetas y leyendas para mejorar la presentación.

## Sección 1: Gráficos simples (líneas, barras, histogramas)

En esta sección aprenderás a crear los tres tipos de gráficos más comunes: líneas, barras e histogramas. Usaremos `pyplot`, un submódulo de Matplotlib que facilita la generación de gráficos con pocos comandos.

```python
import matplotlib.pyplot as plt

#Datos de ejemplo
ventas = [120, 135, 150, 165, 190]
meses = ["Ene", "Feb", "Mar", "Abr", "May"]
```

```python
#Gráfico de líneas
plt.plot(meses, ventas)
plt.title("Ventas mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas (€)")
plt.grid(True)
plt.show()
```

```python
#Gráfico de barras
plt.bar(meses, ventas, color='orange')
plt.title("Ventas mensuales (barras)")
plt.xlabel("Mes")
plt.ylabel("Ventas (€)")
plt.show()
```

```python
#Histograma
edades_clientes = [22, 35, 42, 25, 33, 22, 45, 36, 25, 40, 29, 34]
plt.hist(edades_clientes, bins=5, color='green')
plt.title("Distribución de edades de clientes")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

## Sección 2: Etiquetas y leyendas

Para que los gráficos sean comprensibles, es importante añadir etiquetas y leyendas que indiquen claramente qué representa cada línea o barra. Aquí aprenderás a hacerlo.

```python
# Comparación de dos productos
producto_a = [120, 130, 145, 160, 180]
producto_b = [100, 110, 120, 140, 155]

plt.plot(meses, producto_a, label="Producto A", marker='o')
plt.plot(meses, producto_b, label="Producto B", marker='x')
plt.title("Comparación de ventas")
plt.xlabel("Mes")
plt.ylabel("Ventas (€)")
plt.legend()
plt.grid(True)
plt.show()
```

* Histograma para observar la distribución de los flujos

Se recomienda que experimentes utilizando distintos atributos para generar las gráficas.

```python
#EJECUTA ESTA CELDA PARA PODER UTILIZAR LAS LIBRERÍAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)  # Para reproducibilidad

ventas = np.random.normal(loc=5000, scale=1000, size=120)
```

```python
plt.hist(ventas, bins=10, edgecolor='black', color='skyblue')
plt.title("Distribución de ventas mensuales")
plt.xlabel("Ventas mensuales (€)")
plt.ylabel("Frecuencia")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#1
ingresos = [25000, 27000, 30000, 32000]
trimestres = ["Q1", "Q2", "Q3", "Q4"]
plt.plot(trimestres, ingresos, marker='s')
plt.title("Ingresos Trimestrales")
plt.xlabel("Trimestre")
plt.ylabel("Ingresos (€)")
plt.grid(True)
plt.show()

#2
productos = ["Producto A", "Producto B", "Producto C", "Producto D"]
ventas_prod = [340, 280, 400, 360]
plt.bar(productos, ventas_prod, color="purple")
plt.title("Ventas por producto")
plt.xlabel("Producto")
plt.ylabel("Unidades vendidas")
plt.show()

#3
precios = [15, 20, 22, 30, 25, 18, 35, 40, 22, 28]
plt.hist(precios, bins=5, color="skyblue")
plt.title("Distribución de precios")
plt.xlabel("Precio (€)")
plt.ylabel("Frecuencia")
plt.show()

#4
año_2023 = [12000, 15000, 17000, 16000]
año_2024 = [13000, 15500, 18000, 17500]
plt.plot(trimestres, año_2023, label="2023", linestyle="--", color="blue")
plt.plot(trimestres, año_2024, label="2024", linestyle="-", color="red")
plt.title("Comparativa de beneficios")
plt.xlabel("Trimestre")
plt.ylabel("Beneficio (€)")
plt.legend()
plt.grid(True)
plt.show()

#5
plt.plot(trimestres, ingresos, color="darkgreen", marker="D", linestyle=":")
plt.title("Ingresos personalizados")
plt.xlabel("Trimestre")
plt.ylabel("Ingresos (€)")
plt.grid(True)
plt.show()
```
